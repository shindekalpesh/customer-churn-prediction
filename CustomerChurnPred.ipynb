```python
df.head()
```

```python
# Converting the Target variable:

df['Churn'] = df['Churn'].replace({
    "Yes": 1,
    "No":0
})
```

```python
df['Churn'].value_counts()
```

```python
df.head()
```

```python
# Model Building
```

```python
# Splitting the data into X: Features & y: Target

X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.25)
```

```python
#for i in [X_train,X_test,y_train,y_test]:
#    print(df[i].head(), end="\n")
```

```python
X_train
```

```python
y_train
```

```python
# Implementing LogisticRegression:

lr_clf = LogisticRegression()
lr_model = lr_clf.fit(X_train,y_train)
lr_y_pred = lr_model.predict(X_test)
lr_accuracy_score = round(accuracy_score(y_test,lr_y_pred)*100,2)
print(f"Accuracy of Logistic Regression Classifier is: {lr_accuracy_score}%")
```

```python
# Implementing RandomForest:

rf_clf = RandomForestClassifier()
rf_model = rf_clf.fit(X_train,y_train)
rf_y_pred = rf_model.predict(X_test)
rf_accuracy_score = round(accuracy_score(y_test,rf_y_pred)*100,2)
print(f"Accuracy of Random Forest Tree Classifier is: {rf_accuracy_score}%")
```

```python
# Implementing DecisionTree:

dt_clf = DecisionTreeClassifier()
dt_model = dt_clf.fit(X_train,y_train)
dt_y_pred = dt_model.predict(X_test)
dt_accuracy_score = round(accuracy_score(y_test,dt_y_pred)*100,2)
print(f"Accuracy of Decision Tree Classifier is: {dt_accuracy_score}%")
```
